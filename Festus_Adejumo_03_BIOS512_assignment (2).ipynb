Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library (scales)
```

```r
options(repr.plot.width=5, repr.plot.height=5, repr.plot.res = 300)
p = ggplot (penguins, aes(x=body_mass_g)) +
    geom_histogram(na.rm = T, bins = 25, color = '#3457D5', fill = '#89CFF0') +
    facet_wrap (~year, ncol=1, scale ='free_y') +
    scale_x_continuous(labels = comma_format()) +
    theme_minimal() +
    theme(axis.title =  element_text(size = 10), plot.title = element_text (size =8), 
         plot.title.position = 'panel') +
    labs(x= 'Body Mass in Grams', y = 'Number of Penguins', 
         title = 'Histograms of Penguin Body Mass for Measurements taken from 2007 - 2009')
    
    
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
p = ggplot (penguins, aes(x=body_mass_g, y = sex)) +
    geom_point(na.rm = T, position = position_jitter(h = 0.2, w = 0.15), size = 0.75, alpha = 0.5) +
    facet_wrap(~species, ncol=1) +
    theme_minimal() +
    scale_x_continuous(labels = comma_format()) +
    theme_minimal() +
    theme(axis.title =  element_text(size = 10), plot.title=element_text(hjust = 3) ) +
    labs(x= 'Body Mass in Grams', y ='')
p
    
```
